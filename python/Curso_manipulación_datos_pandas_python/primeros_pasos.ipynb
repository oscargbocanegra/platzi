**Series e Indexación y selección de datos (3/29)**

```python
#Importando libreria pandas
import pandas as pd

# Creando primer elemento series que cuenta con una columna de indexacion
sr = pd.Series([10,9,8,7,6])
print(f'Creando la serie {sr}')
```

```python
# Obteniendo los valores de la serie
print(f'Obteniendo los valores de la serie {sr.values}')
# Obteniendo el indice de la serie
print(f'Obteniendo el indice de la serie {sr.index}')
# Obtener los elementos de la lista 
print(f'Obtener los elementos [0,4,2] de la lista \n{sr[[0,4,2]]}')
```

```python
# Cambiando el sistema de indices de numeros a letras.
sr_alf = pd.Series([10,9,8,7,6], index = ['a','b','c','d','e'])
sr_alf
```

```python
# Se pueden obtener los datos de la serie de diferentes formas por los indices.
# Obteniendo varios elementos.
elemento = sr_alf[['c','d','e']]
print (f'Obteniendo varios elementos por los indices [c,d,e] \n{elemento}')
```

```python
# Obteniendo varios elementos por un rango.
rango = sr_alf['b':'e']
print (f'Obteniendo varios elementos por los indices [b:e] \n{rango}')
```

```python
# Los diccionarios se definen por llaves {} y contienen clave y valor separados por dos puntos :
diccionario = {'CO':100 ,'MX':200 ,'AR':300 }
print(f'Imprimiendo diccionario {diccionario}')
```

```python
# Para extraer las llaves de un diccionario se usa .keys()
print(f'Imprimiendo las llaves del diccionario {diccionario.keys()}')
```

```python
# Para extraer el dato contenido en una de las llaves se usa las llaves con el valor de la clave
llave = diccionario['MX']
print(f'Imprimiendo el valor de la llave MX del diccionario {llave}')
```

```python
# Transformando un diccionario en una serie de pandas.
serie_transformada = pd.Series(diccionario)
print(f'Transformando un diccionario en una serie de pandas. \n{serie_transformada}')
```

```python
# Creando una serie con objetos NaN
serieNaN = pd.Series(serie_transformada, index = ['CO','MX','PE'])
print(f'Creando una serie con objetos NaN \n{serieNaN}')
```

```python
# NaN es un objeto nulo que cumple con operaciones matematicas.
# Cualquier operacion que se realice con NaN su resultado es NaN.
# Aplicando isnull a la serie para validar los objetos verdaderos.
print(f'Con isnull muestra objetos nulos en Falso. \n{serieNaN.isnull()}')
```
